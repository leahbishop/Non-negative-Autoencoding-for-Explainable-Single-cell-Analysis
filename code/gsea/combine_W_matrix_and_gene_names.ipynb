```python
import pandas as pd

W = pd.read_csv(
    "/mnt/projects/debruinz_project/bisholea/capstone/gsea/ae_adata_var_embeddings_unique.csv"
)

meta = pd.read_csv(
    "/mnt/projects/debruinz_project/bisholea/capstone/gsea/ae_adata_var_metadata_unique.csv"
)

print("=== W matrix ===")
print("shape:", W.shape)
print("columns:", list(W.columns)[:10], "...")

print("\n=== Metadata ===")
print("shape:", meta.shape)
print("columns:", list(meta.columns))
```

```python
assert W.shape[0] == meta.shape[0], "Row counts don't match!"

combined = pd.concat([meta.reset_index(drop=True),
                      W.reset_index(drop=True)], axis=1)

combined.to_csv(
    "/mnt/projects/debruinz_project/bisholea/capstone/gsea/nnae_60M_W_with_gene_name.csv",
    index=False
)

print("W:", W.shape, "meta:", meta.shape, "combined:", combined.shape)
print(combined.head())
```
